FUNCTIONS

```python
def mean(a):
  sum=0
  count=0
  for n in a:
    sum+=n;
    count+=1
  avg= sum/count
  return avg
```

```python
def sum(a):
  sum=0
  for n in a:
    sum+=n
  return sum
```

```python
def maximumize(a):
  max=a[0]
  for n in a:
    if(max<n):
      max=n
  return max
```

```python
def minimize(a):
  min=a[0]
  for n in a:
    if(min<n):
      min=n
  return min
```

```python
def normalize(x,y):#normalizing x in terms of y
  maxx=max(x)
  maxy=max(y)
  
  for i in x:
    i=(i/maxx)*maxy
  return x
```

```python
def corelation(x,y):
  x_bar=mean(x)
  y_bar=mean(y)
  nt=0
  x1=y1=0
  for num,n in zip(x,y):
    nt=nt+(num-x_bar)*(n-y_bar)
    x1=x1+(num-x_bar)*(num-x_bar)
    y1=y1+(n-y_bar)*(n-y_bar)
  dt=(x1*y1)**(1/2)
  corr=nt/dt
 # print(f"Correlelation through calculation is {corr}")
  return corr
```

# ***CONFUSION MATRIX***

```python
import numpy as np

def comp_confmat(actual, predicted):

    # extract the different classes
    classes = np.unique(actual)

    # initialize the confusion matrix
    confmat = np.zeros((len(classes), len(classes)))

    # loop across the different combinations of actual / predicted classes
    for i in range(len(classes)):
        for j in range(len(classes)):

           # count the number of instances in each combination of actual / predicted classes
           confmat[i, j] = np.sum((actual == classes[i]) & (predicted == classes[j]))

    return confmat

# sample data
actual = [1, 3, 3, 2, 5, 5, 3, 2, 1, 4, 3, 2, 1, 1, 2]
predicted = [1, 2, 3, 4, 2, 3, 3, 2, 1, 2, 3, 1, 5, 1, 1]

# confusion matrix
print(comp_confmat(actual, predicted))
confusion_matrix=comp_confmat(actual,predicted)
```

***GRADIENT DESCENT OPTIMIZATION ALGORITHM FOR LINEAAR REGRESSION***

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
class slr_gradient_descent:
  def __init__(self, X, Y):
    self.X=X
    self.Y=Y
    self.b=[0,0]
  
  
  def predict(self, X=np.array([])):
    b=self.b
    Y_pred=np.array([])
    for x in X:
      Y_pred=np.append(Y_pred,float(b[0]+(b[1] *x)))
    return Y_pred
  def update_coeff(self,alpha):
    b=self.b
    Y=self.Y
    X=self.X
    Y_pred=self.predict(self.X)
    m=int(len(Y))
    self.b[0]=self.b[0] - (alpha*((np.sum(Y_pred-Y))/m))
    self.b[1]=self.b[1] - (alpha*((np.sum((Y_pred-Y)*self.X))/m))  


  def comp_cost(self,Y_pred):
    Y=self.Y
    m=len(Y)
    J=(1/(2*m))*(np.sum((Y_pred-Y)**2))
    return J
  def plot(self, Y_pred):
    plt.scatter(self.X, self.Y, color='b')
    plt.plot(self.X, Y_pred, color='g')
    plt.show()
```

***Principal Component Analysis***

```python
import numpy as np
import pandas as pd
```

```python
def PCA(X ,components):
     
    #Standardizing the dataset
    X_std = X - np.mean(X , axis = 0)
     
    #Creating a covariance matrix for the data
    CovMat = np.cov(X_std , rowvar = False)
     
    #Calculating the eigen values and corresponding eigen vectors
    EigenValues , EigenVectors = np.linalg.eigh(CovMat)
     
    #Arraging the eigen values and corresponding eigen values
    arranged_ind = np.argsort(EigenValues)[::-1]
    arranged_eigenvalue = EigenValues[arranged_ind]
    arranged_eigenvectors = EigenVectors[:,arranged_ind]
    #Creating a subset of principal components(eigen values)
    eigenvector_subset = arranged_eigenvectors[:,0:components]
    #Step-6
    X_reduced = np.matmul( X_std, eigenvector_subset )
     
    return X_reduced
```
